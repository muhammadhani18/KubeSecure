```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Preprocessing and Feature Selection
from sklearn.preprocessing import LabelEncoder, RobustScaler
from sklearn.feature_selection import VarianceThreshold  # Optional feature selection
from sklearn.model_selection import train_test_split, KFold

# Keras and Regularization
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, LSTM, Dense, Dropout, MaxPooling1D, BatchNormalization, GaussianNoise, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import L1L2
import warnings
warnings.filterwarnings("ignore")
# ==============================
# 1. Data Loading (Using Chunks)
# ==============================
chunksize = 100000  # Adjust based on your system's capacity
data_chunks = []
for chunk in pd.read_csv("/kaggle/input/kubernetes-ddos-attack-dataset/combined_dataset.csv", chunksize=chunksize):
    chunk.dropna(inplace=True)
    chunk.replace([np.inf, -np.inf], np.nan, inplace=True)
    chunk.dropna(inplace=True)
    data_chunks.append(chunk)
df = pd.concat(data_chunks, ignore_index=True)

# # Visualize initial label distribution
# plt.figure(figsize=(8, 6))
# sns.countplot(x='Label', data=df, palette="viridis")
# plt.title("Initial Label Distribution (0: Benign, 1: Anomaly)")
# plt.xlabel("Label")
# plt.ylabel("Count")
# plt.show()

# ============================================
# 2. Data Cleaning & Preprocessing (Pre-Balancing)
# ============================================
df['id'] = df.index  # Create an 'id' column to track original rows

# Save Timestamp separately with id
timestamp_df = df[['id', 'Timestamp']].copy()
df.drop(columns=['Timestamp'], inplace=True)

# Drop columns with >90% zeros (except 'Label' and 'id')
cols_to_drop = []
for col in df.columns:
    if col not in ['Label', 'id']:
        if (df[col] == 0).mean() > 0.9:
            cols_to_drop.append(col)
if cols_to_drop:
    print("Dropping columns with >90% zeros:", cols_to_drop)
    df.drop(columns=cols_to_drop, inplace=True)

# Check missing values per column
missing_counts = df.isnull().sum()
print("Missing values per feature:\n", missing_counts)
# if (missing_counts > 0).any():
#     plt.figure(figsize=(10, 6))
#     missing_counts[missing_counts > 0].plot(kind='bar', color='coral')
#     plt.title("Missing Values Per Feature")
#     plt.ylabel("Count")
#     plt.show()

# =====================================
# 3. Handling Non-Numeric Columns
# =====================================
non_numeric_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()
if non_numeric_cols:
    print("Non-numeric columns detected:", non_numeric_cols)
    df.drop(columns=non_numeric_cols, inplace=True)

# # Visualize cleaned label distribution
# plt.figure(figsize=(8, 6))
# sns.countplot(x='Label', data=df, palette="magma")
# plt.title("Label Distribution After Cleaning")
# plt.xlabel("Label")
# plt.ylabel("Count")
# plt.show()

# =====================================
# 4. Balance the Dataset by Threshold
# =====================================
# Original counts in your dataset:
#  Label 1: ~60,447 samples (malicious)
#  Label 0: ~4,002,078 samples (benign)
n_samples = 60000
benign_df = df[df['Label'] == 0].sample(n=n_samples, random_state=42)
malicious_df = df[df['Label'] == 1].sample(n=n_samples, random_state=42)
balanced_df = pd.concat([benign_df, malicious_df]).sample(frac=1, random_state=42).reset_index(drop=True)
print("Balanced dataset shape:", balanced_df.shape)
print("Balanced label distribution:\n", balanced_df['Label'].value_counts())

# =====================================
# 5. Reintegrate Timestamp Using id
# =====================================
balanced_df = pd.merge(balanced_df, timestamp_df, on='id', how='left')

# =====================================
# 6. Prepare Dataset for Sequence Generation
# =====================================
# Remove columns that are not needed for training.
for col in ['Flow ID', 'Src IP', 'Dst IP']:
    if col in balanced_df.columns:
        balanced_df.drop(columns=[col], inplace=True)
        
# Encode 'Protocol' if it exists.
if 'Protocol' in balanced_df.columns:
    protocol_le = LabelEncoder()
    balanced_df['Protocol'] = protocol_le.fit_transform(balanced_df['Protocol'])

# Encode the Label column (if not already numeric 0/1).
label_le = LabelEncoder()
balanced_df['Label'] = label_le.fit_transform(balanced_df['Label'])

# Convert 'Timestamp' to numerical value.
if 'Timestamp' in balanced_df.columns:
    balanced_df['Timestamp'] = pd.to_datetime(balanced_df['Timestamp'], errors='coerce')
    balanced_df['Timestamp'] = balanced_df['Timestamp'].astype('int64', errors='ignore') // 10**9
    balanced_df['Timestamp'].fillna(0, inplace=True)

# Drop the 'id' column for training.
if 'id' in balanced_df.columns:
    balanced_df.drop(columns=['id'], inplace=True)

# Separate features and labels.
X = balanced_df.drop(columns=['Label']).values
y = balanced_df['Label'].values


# Normalize features using RobustScaler.
scaler = RobustScaler()
X = scaler.fit_transform(X)

# After preparing your dataset for sequence generation (before creating sequences)
# The features used in training are all columns except 'Label'
feature_names = list(balanced_df.drop(columns=['Label']).columns)
print("Features used for training:")
print(feature_names)

# Save the feature names to a text file for future reference.
with open("/kaggle/working/features_used.txt", "w") as f:
    for feature in feature_names:
        f.write(feature + "\n")
print("Feature names saved to 'features_used.txt'.")


# =====================================
# 7. Create Sequences from the Data
# =====================================
def create_sequences(X, y, seq_length=10):
    X_seq, y_seq = [], []
    for i in range(len(X) - seq_length + 1):
        X_seq.append(X[i:i+seq_length])
        y_seq.append(y[i+seq_length-1])
    return np.array(X_seq), np.array(y_seq)

sequence_length = 10
X_seq, y_seq = create_sequences(X, y, seq_length=sequence_length)
print("Sequence data shape:", X_seq.shape)
print("Sequence labels shape:", y_seq.shape)

# =====================================
# 8. Split the Sequences into Training and Testing Sets
# =====================================
X_train, X_test, y_train, y_test = train_test_split(X_seq, y_seq, test_size=0.2, random_state=42, stratify=y_seq)
print("Training sequence shape:", X_train.shape)
print("Testing sequence shape:", X_test.shape)

# =====================================
# 9. Build the Improved CLSTM Model
# =====================================
input_shape = (X_train.shape[1], X_train.shape[2])
reg = L1L2(l1=1e-5, l2=1e-4)  # Regularization settings

model = Sequential()
# Input layer with Gaussian noise for robustness
model.add(Input(shape=input_shape))
model.add(GaussianNoise(0.1))
model.add(BatchNormalization())

# First Conv1D layer with regularization
model.add(Conv1D(filters=32, kernel_size=3, activation='relu', kernel_regularizer=reg))
model.add(BatchNormalization())

# Second Conv1D layer with a different kernel size
model.add(Conv1D(filters=32, kernel_size=5, activation='relu', kernel_regularizer=reg))
model.add(BatchNormalization())
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.5))

# LSTM layer with reduced units
model.add(LSTM(64, kernel_regularizer=reg))
model.add(BatchNormalization())
model.add(Dropout(0.5))

# Fully connected layers with regularization
model.add(Dense(32, activation='relu', kernel_regularizer=reg))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid', kernel_regularizer=reg))  # Binary classification

optimizer = Adam(learning_rate=0.0005)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# =====================================
# 10. Callbacks for Training
# =====================================
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6, verbose=1)

# =====================================
# 11. Model Training (Option 1: Simple Train/Test Split)
# =====================================
history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=16,
    validation_split=0.2,
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)


# =====================================
# 12. Visualize Training Performance
# =====================================
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss', color='blue')
plt.plot(history.history['val_loss'], label='Val Loss', color='orange')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training and Validation Loss")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train Acc', color='blue')
plt.plot(history.history['val_accuracy'], label='Val Acc', color='orange')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training and Validation Accuracy")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix

# Plot training and validation loss
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Evaluate the model on the test data
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=1)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")

# Make predictions on the test set
y_pred = (model.predict(X_test) > 0.5).astype("int32")

# Classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Benign', 'Anomaly'], yticklabels=['Benign', 'Anomaly'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()


# Save the model
model.save('/kaggle/working/clstm_anomaly_detection_model.h5')
print("Model saved as 'clstm_anomaly_detection_model.h5'")
```
